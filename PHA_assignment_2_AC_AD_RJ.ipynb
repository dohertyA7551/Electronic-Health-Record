```python
#SECTION 1

#defining the person class using the constructor
class Person:
    def __init__(self, person_ID, birth_year, contact_no):
        self.__id = person_ID #encapsulating the ID using double underscores
        self.yob = birth_year
        self.contact = contact_no

#creating getter methods for each attribute
    def get_id(self):
        return self.__id
    def get_yob(self):
        return self.yob
    def get_contact(self):
        return self.contact

#creating setter methods for each attribute
    def set_id(self, person_ID):
        self.__id = person_ID
    def set_yob(self, birth_year):
        self.yob = birth_year
    def set_contact(self, contact_no):
        self.contact = contact_no

#creating class method to check if person is an adult
    def is_adult(self):
        age = 2026 - self.yob #calculating age
        if age < 18:
            return 'Minor'
        elif 18 <= age <70:
            return 'Adult'
        else:
            return 'Senior Citizen'

#creating class method to print details of the person
    def print_details(self):
        age = 2026 - self.yob #calculating age
        return f'person ID: {self.__id},person age: {age},person contact number {self.contact}'

#example person to make sure code is correct/working
john = Person(123, 2001, "0871234567")
print(john.print_details())
```

```python
#SECTION 2

import Bio
from Bio.Seq import Seq

#new patient class inheriting from person class
#three new parameters for weight, height and empty medical history list
class Patient(Person):
    def __init__(self, person_ID, birth_year, contact_no, weight, height):
        super().__init__(person_ID, birth_year, contact_no)
        self.weight = weight
        self.height = height
        self.medical_history = []

#new function to calculate and return bmi
    def calculate_bmi(self, weight, height):
        self.weight = weight
        self.height = height
        self.bmi_full = self.weight / (self.height ** 2)
        self.bmi_rounded = round(self.bmi_full, 2)
        return self.bmi_rounded

#getter functions for easy handling
    def get_weight(self):
        return self.weight
    def get_height(self):
        return self.height
    def get_medical_history(self):
        return self.medical_history
    def get_bmi_rounded(self):
        return self.calculate_bmi(self.weight, self.height)

#append new entry function for medical history list
    def add_medical_history(self, new_entry):
        self.medical_history.append(new_entry)
        print(f'{new_entry} added to patient medical history: {self.medical_history}')

#removal function of most recent addition to medical history list
    def remove_medical_history(self):
        len_medical_history = len(self.medical_history)
        if len_medical_history > 0:
            print(f'{self.medical_history[-1]} will be removed from patient medical history.')
            remove_item = self.medical_history.remove(self.medical_history[-1])
            print(f'Removed from medical history. Updated version: {self.medical_history}')
        else:
            print('Medical history is empty')

#Analyze a DNA sequence to give back a dictionary containing the sequence, transcribed, translated,GC content
    def analyse_dna_sequence(self, dna_seq):
        sequence = Seq(dna_seq)
        transcribed_seq = sequence.transcribe()
        translated_seq = sequence.translate()
        G_count = sequence.count("G")
        C_count = sequence.count("C")
        GC_percent=(G_count+C_count)/len(dna_seq)*100
        compiled_info = {"DNA sequence":str(dna_seq), "GC content (%)": GC_percent, "transcribed": str(transcribed_seq), "translated": str(translated_seq)}
        return compiled_info

#printing details uses parent class method, followed by additional bmi and history
    def print_details(self):
        person_print_details = super().print_details()
        bmi_value = self.calculate_bmi(self.weight, self.height)
        return(f'{person_print_details} bmi {bmi_value}, medical history: {self.medical_history}')

#test data
steve = Patient(124, 2001, "0871234567", 77, 2)
print(f'{steve.get_bmi_rounded()} units')
print(f'history: {steve.get_medical_history()}')
steve.add_medical_history("shivers")
steve.add_medical_history("sweating")
steve.remove_medical_history()
stacy = Patient(125, 2003, "0871234567", 77, 2)
stacy.remove_medical_history()
patient_seq = input("Enter patient sequence: ")
print(steve.analyse_dna_sequence(patient_seq))
stacy.add_medical_history("bleeding")
print(stacy.print_details())
```

```python
#SECTION 3

#making the third class inheriting from the patient class
class InHospitalPatient(Patient):
    def __init__(self, person_ID, birth_year, contact_no, weight, height, doctor_name, room_no, days_admit):
        super().__init__(person_ID,birth_year,contact_no,weight,height)
        self.doctor_name = doctor_name
        self.room_no = room_no
        self.days_admit = days_admit

#method to count_admitted_days
    def count_admitted_days(self, days_admit):
        self.days_admit += 1
        return(f'{self.days_admit} days admitted')

#method collect info on admitted patient
    def admit_patient(self,doctor_name,room_no, days_admit):
        return (f'admitted patient: {self.doctor_name}, {self.room_no}, {self.days_admit}')

#method for when patient is discharged
    def discharge_patient(self):
        self.doctor_name = 0
        self.room_no = 0
        self.days_admit = 0
        return (f'discharged patient: {self.doctor_name}, {self.room_no}, {self.days_admit}')

#print details method specific for this class
    def print_details(self):
        if self.days_admit > 0:
            return( f'{super().print_details()}, Doctor name: {self.doctor_name}, room no: {self.room_no}, total days:{self.days_admit}')
        else:
            return(f'{super().print_details()}')

#test data
robin = InHospitalPatient(124, 2001, "0871234567", 77, 3, "john", 101, 0)
print(f'{robin.count_admitted_days(0)}')
print(f'{robin.admit_patient("john", 101, 0)}')
print(f'{robin.discharge_patient()}')
print(f'{robin.print_details()}')
```
